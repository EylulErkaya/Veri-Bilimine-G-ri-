```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings("ignore")
veri = pd.read_csv("olimpiyatlar_madalya_alanlar.csv")
```

```python
veri.head()
```

```python
def plotHistogram(degisken):
    plt.figure()
    plt.hist(veri[degisken], bins = 85, color = "orange")
    plt.xlabel(degisken)
    plt.ylabel("frekans")
    plt.title(f"{degisken} - Histogram Grafigi")
    plt.show()
```

```python
plotHistogram("yas")
```

```python
sayisal_degiskenler = ["yas", "yil", "boy", "kilo"]
for elem in sayisal_degiskenler:
    plotHistogram(elem)
```

```python
#kategorik degiskenler için
def plotbar(degisken, n = 5):
    veri_sayma = veri[degisken].value_counts()
    print(veri[degisken].value_counts())
    veri_sayma = veri_sayma[:n]
    plt.figure()
    plt.bar(veri_sayma.index, veri_sayma, color = "blue")
    plt.xticks(veri_sayma.index, veri_sayma.index.values)
    plt.xticks(rotation = 90)
    plt.ylabel("frekans")
    plt.title(f"{degisken} - Histogram Grafiği")
    plt.show()
    print(f"{degisken} Analizi: \n(veri_sayma)")
    
    
    
# plotbar("isim", 10)
kategorik_degiskenler = ["isim", "cinsiyet", "takim", "madalya", "sezon", "sehir", "etkinlik"]
for e in kategorik_degiskenler:
    plotbar(e)
```
